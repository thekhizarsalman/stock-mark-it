```python
import pandas
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2020,4,17)
```

```python
google = web.DataReader ("GOOGL","yahoo",start,end)
```

```python
google.head()
```

```python
google ['Open']
```

```python
google ['Open'].plot()
plt.show
```

```python
google ['Open'].plot(label = 'GOOGL Open price')
google ['Close'].plot(label = 'GOOGL Close price')
google ['High'].plot(label = 'GOOGL High price')
google ['Low'].plot(label = 'GOOGL Low price')
plt.legend()
plt.title('google stock price')
plt.show
```

```python
google['Volume'].plot(figsize=(17,5))
plt.title('no. of shares')
plt.show
```
